class sklearn.tree.DecisionTreeClassifier(*, criterion='gini', splitter='best', max_depth=None, min_samples_split=2, min_samples_leaf=1, min_weight_fraction_leaf=0.0, max_features=None, random_state=None, max_leaf_nodes=None, min_impurity_decrease=0.0, class_weight=None, ccp_alpha=0.0, monotonic_cst=None)

Explanations :
1. **criterion:** The function to measure the quality of a split. â€œginiâ€, â€œentropyâ€, â€œlog_lossâ€
2. **splitter:** The strategy used to choose the split at each node. â€œbestâ€, â€œrandomâ€
3. **max_depth:** The maximum depth of the tree.
4. **min_samples_split:** The minimum number of samples required to split an internal node
5. **min_samples_leaf:** The minimum number of samples required to be at a leaf node.
6. **min_weight_fraction_leaf:**The minimum weighted fraction of the sum total of weights required to be at a leaf node.
7. **max_features:** The number of features to consider when looking for the best split
8. **random_state:** Controls the randomness of the estimator. The features are always randomly permuted at each split, even if splitter is set to "best".
9. **max_leaf_nodes:** Grow a tree with max_leaf_nodes in best-first fashion
10. **min_impurity_decrease:** A node will be split if this split induces a decrease of the impurity greater than or equal to this value.
11. **class_weight:** Weights associated with classes in the form {class_label: weight}(dictionary). If None, all classes are supposed to have weight one. For multi-output problems, a list of dicts can be provided in the same order as the columns of y.
12. **ccp_alpha:** Complexity parameter used for Minimal Cost-Complexity Pruning.
13. **monotonic_cst:** Indicates the monotonicity constraint to enforce on each feature.

To determine the optimal values for the hyperparameters of DecisionTreeClassifier I use  **GridSearchCV**.exploring a range of hyperparameter values and find the combination that yields the best performance.

Approach:
1. **Starting with a broad range:** choosing a broad range for each hyperparameter to get an understanding of the general behavior of the model.
2. **Narrowing down the range:** Based on the initial results, narrow down the ranges to focus on the most promising values.



class sklearn.model_selection.GridSearchCV(estimator, param_grid, *, scoring=None, n_jobs=None, refit=True, cv=None, verbose=0, pre_dispatch='2*n_jobs', error_score=nan, return_train_score=False)

**Explanations :**
1. **estimator: ** data to be estimated
2. **param_grid:** Dictionary with parameters names as keys and lists of parameter settings to try as values, or a list of such dictionaries. This enables searching over any sequence of parameter settings.
3. **scoring:** Strategy to evaluate the performance of the cross-validated model on the test set.
4. **n_jobs:** Number of jobs to run in parallel.
5. **refit:** Refit an estimator using the best found parameters on the whole dataset.
6. **cv:** Determines the cross-validation splitting strategy.
7. **verbose:** Controls the verbosity: the higher, the more messages.
8. **pre_dispatch:** Controls the number of jobs that get dispatched during parallel execution. Reducing this number can be useful to avoid an explosion of memory consumption when more jobs get dispatched than CPUs can process.
9. **error_score:** Value to assign to the score if an error occurs in estimator fitting.
10. **return_train_score:** Computing training scores is used to get insights on how different parameter settings impact the overfitting/underfitting trade-off. However computing the scores on the training set can be computationally expensive and is not strictly required to select the parameters that yield the best generalization performance.




```python
# q9: new other models - classification models
# DecisionTreeClassifier : searching the best hyperparamters over a range of inputs
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

# Data without the failure features
x_train_no_failure_features = x_train.drop(["TWF","HDF","PWF","OSF","RNF"], axis = 'columns')
x_test_no_failure_features = x_test.drop(["TWF","HDF","PWF","OSF","RNF"], axis = 'columns')

# DecisionTreeClassifier
DTC = DecisionTreeClassifier()

# Stratified 5-fold cross validation
k_fold_cv = 5
grid_params = {"criterion":["gini","entropy"], "max_depth":[None,5,10,15]}
grid = GridSearchCV(DTC, param_grid = grid_params, cv = k_fold_cv, n_jobs = 1, verbose = 0, return_train_score=True)

grid.fit(x_train_no_failure_features, y_train)
print("Best hyper parameter:", grid.best_params_)
```
